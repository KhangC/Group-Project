```python
crime_df = pd.read_csv("../Project1/Crime.csv")
crime_df.head()
```

```python
print (crime_df.nlargest(3,'violent_crime'))
```

```python
print (crime_df.nsmallest(3,'violent_crime'))
```

```python
crime_df_Alameda_County = crime_df[crime_df.geo_name == "Alameda County, CA"]
crime_df_Alameda_County = crime_df_Alameda_County.sort_values('year')
crime_df_Alameda_County
```

```python
x_axis = crime_df_Alameda_County.year
y_axis = crime_df_Alameda_County.violent_crime
plt.xticks(ticks=x_axis)
plt.grid(True)
plt.plot(x_axis,y_axis)

plt.show()
```

```python
income_df = pd.read_csv("../Project1/Income.csv")
income_df.head()
```

```python
income_df_Alameda_County = income_df[income_df.geo_name == "Alameda County, CA"]
income_df_Alameda_County = income_df_Alameda_County.sort_values('year')
income_df_Alameda_County
```

```python
x_axis2 = income_df_Alameda_County.year
y_axis2 = income_df_Alameda_County.income
plt.xticks(ticks=x_axis2)
plt.grid(True)
plt.plot(x_axis2,y_axis2)
plt.show()
```

```python
crime_income_df = pd.merge(crime_df,income_df,how="left",on=["year","geo"])
crime_income_df
del crime_income_df["geo_name_y"]
crime_income_df = crime_income_df.sort_values(by=["geo_name_x","year"])
crime_income_df["geo_name_x"] = crime_income_df["geo_name_x"].map(lambda x: x.rstrip(', CA'))
crime_income_df.head(10)
```

```python
#print (crime_income_df.sort_values('violent_crime',ascending=False))
print (crime_income_df.nlargest(10,'violent_crime'))
```

```python
sat_df = pd.read_csv('../Project1/Sat.csv')
sat_df.head()
```

```python
sat_df.columns
```

```python
sat_df = sat_df[['Name','Percent Meeting ELA Benchmarks','Percent Meeting Math Benchmarks']]
sat_df = sat_df[sat_df.Name != 'Statewide']
sat_df.head()
```

```python
sat_df = sat_df.sort_values('Percent Meeting ELA Benchmarks',ascending=False)
sat_df.head()
```

```python
sat_df = sat_df.sort_values('Percent Meeting Math Benchmarks',ascending=False)
sat_df.head()
```

```python
sat_df = sat_df.replace('*', np.NaN)
sat_df
```

```python
sat_df['Percent Meeting ELA Benchmarks'].str.rstrip('%').astype('float')
```

```python
sat_df['Percent Meeting Math Benchmarks'].str.rstrip('%').astype('float')
sat_df['Percent Meeting ELA Benchmarks'].str.rstrip('%').astype('float')
```
