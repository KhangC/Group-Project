# Import Dependencies

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Reference to Crime CSV and reading CSV into Pandas DataFrame

```python
df = pd.read_csv('Resources/final_score.csv')
df2 = df.copy()
df2 = df2.sort_values('Score',ascending=False)
df2 = df2.set_index('County')
df = df.sort_values('Score')

df = df.fillna(0)
df = df.drop(['Score'],axis=1)
df.head()
df2
```

# Plot

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

ax.barh(df.County,df.Crime + df.Education + df.Housing,color='purple',align='center',label='Best Housing')
ax.barh(df.County,df.Crime + df.Education,color='gold',align='center',label='Best Education')
ax.barh(df.County,df.Crime, color='pink',align='center',label='Less Crime')

plt.legend()
plt.title('Crime Education Housing Score')
plt.xlabel('Score')
plt.ylabel('Counties')

plt.savefig('Crime_Edu_housing_bar_Chart.png')
plt.show()
```
